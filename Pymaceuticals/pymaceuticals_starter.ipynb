### Analysis

- After cleaning the data both Ramicane and Capomulin stood out as potentially effective drugs because they have the lowest final tumor volume.

- Based on the violin and box plots, both Ramicane and Capomulin performed better at reducing tumor size when compared with the placebo. Additionally, they performed very similarly to eachother.

- When comparing mouse weight to tumor size, a weak positive correlation was found for rats taking Capomulin

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import seaborn as sns

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
df_max_time = df_clean.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df_tumor = pd.merge(df_max_time, df_clean, on=["Mouse ID", "Timepoint"], how="left")

df_tumor.head()
```

```python
#limiting the list of drugs to the 4 most promising and Placebo
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin", "Placebo"]

df_drug_tumor = df_tumor.loc[df_tumor["Drug Regimen"].isin(drugs)].reset_index(drop=True)

```

```python
# Change Canvas Size
plt.figure(figsize=(12, 6))

# Create a basic plot
sns.boxplot(data=df_drug_tumor, y="Tumor Volume", x="Drug Regimen", color="skyblue")

# Customizations

# Change colors
# Add in reference lines
# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Arial", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Tumor Volume by Drug", fontweight="bold", fontsize=16, fontfamily="Arial")

# Add in a background grid
plt.grid(axis="y", linestyle="-", color="lightgrey", alpha=0.75)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = df_tumor["Drug Regimen"].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumor_vols = []


# Calculate the IQR and quantitatively determine if there are any potential outliers.
#loop through the regimens
for treatment in drugs:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_data = df_tumor.loc[df_tumor["Drug Regimen"] == treatment, "Tumor Volume"]

    # add subset
    tumor_vols.append(treatment_data)

    #find Q1 and Q3
    q1 = np.percentile(treatment_data, 25)
    q3 = np.percentile(treatment_data, 75)

    #calculate iqr
    iqr = q3 - q1
    # Determine outliers using upper and lower bounds
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)

    outliers = treatment_data[(treatment_data<lower) | (treatment_data>upper)]
    print(f"Outliers for {treatment}: {outliers.tolist()}")

```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.boxplot(tumor_vols, labels=drugs)

# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Arial", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Tumor Volume by Drug", fontweight="bold", fontsize=16, fontfamily="Arial")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="lightgrey", alpha=0.75)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = weight
y_values = tumor_avg

# do regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept # these are the points/predictions
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
data = df_clean.loc[(df_clean["Drug Regimen"] == "Capomulin")]
tumor_avg = data.groupby("Weight (g)")["Tumor Volume"].mean()
weight = data["Weight (g)"].unique()


# Change Canvas Size
plt.figure(figsize=(12, 6))

#make a basic plot
plt.scatter(weight, tumor_avg, color = "purple")
plt.plot(weight, regress_values, color="black", linewidth=6)

# Add Labels/Title
plt.xlabel("Weight (g)", fontfamily="Arial", fontsize=12)
plt.ylabel("Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Average Tumor Size by Mouse Weight", fontweight="bold", fontsize=16, fontfamily="Arial", fontstyle="italic")




# Show/Save the Graph
plt.show()
```
